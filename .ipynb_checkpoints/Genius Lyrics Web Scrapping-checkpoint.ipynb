```python
def get_link(artist, track):
    artist = re.findall("\w+",artist)
    if len(artist)==1: 
        artist = artist[0][0].upper()+artist[0][1:].lower()
    else:
        artist = artist[0]+"-"+"-".join(artist[1:])
    track = "-".join(re.findall("\w+",track.lower()))
    url = "https://genius.com/"+artist+"-"+track+"-lyrics"
    return url

def make_request(link): 
    try:
        r = requests.get(link)
        if r.status_code==200:
            return r.text
        else: 
            return ""
    except requests.exceptions.HTTPError as e:
        print(e.response.text)
        return ""

def get_lyric(artist,track): 
    link = get_link(artist, track)
    html = make_request(link)
    return html
```

```python
page_text = get_lyric("21 Savage","Glock In My Lap")
soup = BeautifulSoup(page_text, 'html.parser')
print(soup.prettify())
```
